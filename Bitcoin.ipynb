```
import csv
import matplotlib.pyplot as plt
filename="bitcoin.csv"
fields=[] 
rows=[]
dates=[]
opening=[]
closing=[]
high=[]
low=[]
gross=[]
with open(filename, 'r') as csvfile: 
    csvreader=csv.reader(csvfile) 
    fields=next(csvreader) 
    for row in csvreader: 
        rows.append(row)
for row in rows[:90]:
    dates.append(row[0])
    opening.append(row[1])
    high.append(row[2])
    low.append(row[3])
    closing.append(row[4])
opening=list(map(float,opening))
closing=list(map(float,closing))
high=list(map(float,high))
low=list(map(float,low))
'''print(dates)
print(opening)
print(closing)
print(high)
print(low)'''
for i in range(len(opening)):
  gross.append(opening[i]-closing[i])
x=dates
y=gross
xerr=[0.01]*len(dates)
yerr=[0.001]*len(gross)

plt.rc('font',family='serif',size=13)
plt.plot(x,y,color='#1076FF')
plt.show()
```
